# Traffic Safety

```python
A = 13 #number of crashes 
num = A*1e6 # numerator
adt = [9671,2893,9506,2611] # we are going to store the ADT values in a list 
tot = sum(adt) 
year = 365 # 1 year
den = 0.5*tot*year # denominator. we are going to divide the total by 2 because this is a 2-way ADT
rmev = num/den # calculate the rmev
print(f'RMEV: {round(rmev,2)} crashes per million')
```

```python
L = 6.1 # miles (length of the road)
A = 6 # number of accidents
adt = 755 # vehicles per day 
days = 365 # 1 year 
vmt = adt*days*L 
rhmv = (A*1e8)/vmt
print(f'RHMV: {round(rhmv)} crashes per hundred million vehicle miles')
```

```python
crs = [40,28,20] # 3 countermeasures
crs = [i/100 for i in crs] # convert them to decimals
cr_tot = 0 # we are going to add the crs for each location
for i in range(len(crs)):
    temp = 1
    for j in range(0,i):
        temp *= (1-crs[j])
    cr_tot += temp*crs[i]

print(f'cr total: {cr_tot}')
```

```python
crashes = [12,14,13]
N = np.mean(crashes)
adt_before = 7850 # adt before
adt_after = 9000 # adt after
crashes_prevent = N*cr_tot*(adt_after/adt_before)
print(f'Crashes prevented: {round(crashes_prevent,2)}')
```
