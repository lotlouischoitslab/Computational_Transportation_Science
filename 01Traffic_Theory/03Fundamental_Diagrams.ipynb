# Fundamental Diagrams

There are 3 fundamental diagrams transportation engineers must know

### Greenshields Model

$$v(k) = v_f(1-\frac{k}{k_j})$$

**vf: Means the free-flow speed (or the speed where the density k = 0)**

**kj: Jam density basically means the density where vf = 0**

**kc: Critical Density is the density where the traffic flow q is at maximum)**

```python
k = np.linspace(0,10,100)
vf = 5
kj = 10
v = vf*(1-(k/kj))
plt.plot(k,v)
plt.xlim(left=0)
plt.xlim(right=10)
plt.ylim(bottom=0)
plt.xlabel('Density k')
plt.ylabel('Speed v')
plt.title('Speed vs Density')
```

```python
k = np.linspace(0,10,100) # range of density values
vf = 5 # free flow speed
kj = 10 # jam density
v = vf*(1-(k/kj)) # Greenshield's model
plt.plot(k,v) # plot the curve

kc = k[len(k)//2] # critical density
vc = v[len(v)//2] # max flow speed

# Define the range of x and y values for the axvline and axhline
x_vline = np.linspace(0, kc, 10)
y_vline = np.full_like(x_vline, vc)

y_hline = np.linspace(0, vc, 10)
x_hline = np.full_like(y_hline, kc)

plt.plot(x_vline, y_vline, linestyle='--', label='Max flow speed vc', color='b')
plt.plot(x_hline, y_hline, linestyle='--', label='Critical Density kc', color='r')

plt.xlim(left=0)
plt.xlim(right=10)
plt.ylim(bottom=0)
plt.xlabel('Density k (veh/km)')
plt.ylabel('Speed v (km/hr)')
plt.title('Speed vs Density')
plt.legend()
```

```python
k = np.linspace(0, 10, 100)
vf = 5 # free-flow speed
kj = 10 # jam density
q = vf * k * (1 - (k / kj))
plt.plot(k,q)
plt.xlabel('Density k (veh/km)')
plt.ylabel('Flow q (veh/hr)')
plt.title('Flow vs Density')
```

# Triangular Fundamental Diagram

```python
k = np.array([0,5,10,15,20,25,30,35,40,45,50,70,120])
vf = 5
w = 1
kc = 20
index_of_kc = list(k).index(kc)
q1 = vf*k
q2 = -w*(k-kc)+vf*kc
plt.plot(k[:index_of_kc+1],q1[:index_of_kc+1],label='Uncongested',color='blue')
plt.plot(k[index_of_kc:],q2[index_of_kc:],label='Congested',color='red')
plt.ylim(0)
plt.xlabel('Density veh/km')
plt.ylabel('q (veh/hr)')
plt.title('Triangular Fundamental Diagram')
plt.axvline(kc,linestyle='--',label='Critical Density kc')
plt.plot(120,q2[len(q2)-1],'bo',label='Jam Density kj')
plt.legend()
```

```python
# Part 1
from sympy import * 
k = symbols('k') # We need to calculate the density
v = 60-0.2*k # Greenshield's model
m = 45 # mean speed on the road in mph
equation = m - v
sol = solve(equation,k)
k_ans = sol[0]
print(f'Density: {k_ans} vehicles per mile')
```

```python
# Part 2 
from sympy import * 
k = symbols('k') # We need to calculate the density
v = 60-0.2*k # Greenshield's model
vc = 30 # max flow speed
equation = m - v # set up our equation
sol = solve(equation,k) # solve the equation
kc_ans = sol[0] # Store this as our critical density kc
print(f'Critical Density: {k_ans} vehicles per mile')
q = kc_ans*vc # Traffic Flow Theory Formula 
print(f'Maximum Flow: {q} vehicles per hour')
```
