# Trajectory Diagrams
Best way to visualize traffic movements in terms of distance and time.

```python
t = np.linspace(1,10)
y = (t-5)**3+100
plt.scatter(t,y,color='y',marker='*',label='Vehicle')
plt.plot(t,y,label='Trajectory Curve')
plt.xlabel('Time in seconds')
plt.ylabel('Distance in x(t)')
plt.title('Trajectory Diagram')
plt.ylim(0)
plt.legend()
```

**Each point along the trajectory curve represents a vehicle movement. Remeber, this one curve is for one vehicle.**

```python
t = np.linspace(1,10)
y = (t-5)**3+100
plt.plot(t,y,label='Trajectory Curve')
plt.plot(t[5:len(t)//2- 3],y[5:len(t)//2 - 3],label='Braking',color='red')
plt.plot(t[len(t)//2- 3:-6],y[len(t)//2 - 3:-6],label='Accelerating',color='green')
plt.xlabel('Time in seconds')
plt.ylabel('Distance in x(t)')
plt.title('Trajectory Diagram')
plt.ylim(0)
plt.legend()
```

```python
t = np.linspace(1,10)
counter = 1
for i in range(20,100,10):
    y = (t-5)**3+i
    plt.plot(t,y,label='Vehicle N.o. '+str(counter))
    counter+=1
plt.xlabel('Time in seconds')
plt.ylabel('Distance in x(t)')
plt.title('Trajectory Diagram')
plt.ylim(0)
plt.legend()
```

**Trajectory curve for multiple vehicles shown above.**

# Traffic Measurements

**Stationary Observer (SO): Counts the number of vehicles at a fixed location </br>**
**Overhead Observer (OO): Counts the number of vehicles above at a given length**

```python
x = np.linspace(0,10)
m = 5
b = [i for i in range(0,50,2)]
for b in b:
    y = m*x+b
    plt.plot(x,y)
    plt.title('Trajectory')
    plt.xlabel('Time')
    plt.ylabel('Distance')
    plt.xlim(right=8)
    plt.ylim(top=100)
plt.axhline(y = 50, color = 'black', linestyle = '-',label='Stationary T')
plt.legend()
```

Stationary Observers as shown above count the number of vehicles horizontally. Such examples include loop detector, RFID gantors and bluetooth reader.

```python
x = np.linspace(0,10)
m = 5
b = [i for i in range(0,50,2)]
for b in b:
    y = m*x+b
    plt.plot(x,y)
    plt.title('Trajectory')
    plt.xlabel('Time')
    plt.ylabel('Distance')
    plt.xlim(right=8)
    plt.ylim(top=100)
plt.axvline(x = 5, color = 'black', linestyle = '-',label='Overhead k')
plt.legend()
```

Overhead observers as shown above count the number of vehicles from a top view. Such examples are drones, cameras, and traffic signal controllers
