## 2. Python Functions

```python
# Fibonacci Number f[i] = f[i-1] + f[i-2]
def fib(n):
  f = [0 for i in range(n)] # populate the element 0 for n times
  f[0] = 0
  f[1] = 1

  # for i in range(start_index,end_index,increment)

  # for i in range(end_index) automatically go from 0 to end_index and increment by 1
  # for i in range(start_index,end_index) go from start_index to end_index and increment by 1
  # for i in range(start_index,end_index,increment_index)

  # for i in range(end_index,start_index,-1)

  for i in range(2,n):
    f[i] = f[i-1] + f[i-2]
  return f[n-1]
```

```python
answer = fib(8)
answer
```

```python
def fib_new(n):
  f = [0 for i in range(n)] # populate the element 0 for n times
  f[0] = 0
  f[1] = 1

  # for i in range(start_index,end_index,increment)

  # for i in range(end_index) automatically go from 0 to end_index and increment by 1
  # for i in range(start_index,end_index) go from start_index to end_index and increment by 1
  # for i in range(start_index,end_index,increment_index)

  # for i in range(end_index,start_index,-1)

  for i in range(2,n):
    f[i] = f[i-1] + f[i-2]
  return f, f[n-1]   # first_variable , second_variable
```

```python
n = 8
dp, answer = fib_new(n)
print(dp)
print(answer)
```

```python
def count_elem(my_list):
  louis_dict = dict()
  for elem in my_list: # go each element inside the list
    if elem not in louis_dict:
      louis_dict[elem] = 1
    else:
      louis_dict[elem] += 1
  return louis_dict
```

### Linear Optimization

\begin{alignat*}{2}
\text{Minimize:} \quad & -2x - 3y \\
\text{Subject to:} \quad & x + y &\leq 5 \\
                        & x - 2y &\leq 2 \\
                        & -x + y &\leq 3 \\
                        & x &\geq 0 \\
                        & y &\geq 0
\end{alignat*}

```python
from scipy.optimize import linprog
c = [-2,-3]
A = [
    [1,1],
    [1,-2],
    [-1,1]
]

b = [5,2,3]

x_bounds = (0,None)
y_bounds = (0,None)

result = linprog(c,A_ub=A,b_ub=b,bounds=[x_bounds,y_bounds])
result.x # first print out result, if success, then go ahead and print out result.x
```

# Probability and Statistics

## Normal Distribtution

### Probability Density Function (PDF):

$ f(x|\mu,\sigma) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}} $


### Cumulative Distributive Function (CDF):
$ F(x|\mu,\sigma) = \frac{1}{2}\left[1 + \text{erf}\left(\frac{x-\mu}{\sqrt{2\sigma^2}}\right)\right] $

## Uniform Distribution

### Probability Density Function (PDF):

$ f(x|a,b) = \frac{1}{b-a} \quad \text{for } a \leq x \leq b $

### Cumulative Distributive Function (CDF):
$
F(x|a,b) = \begin{cases}
             0 & \text{for } x < a \\
             \frac{x-a}{b-a} & \text{for } a \leq x \leq b \\
             1 & \text{for } x > b
           \end{cases}
$

## Exponential Distribution

### Probability Density Function (PDF):

$ f(x|\lambda) = \lambda e^{-\lambda x} \quad \text{for } x \geq 0 $

### Cumulative Distributive Function (CDF):

$F(x|\lambda) = 1 - e^{-\lambda x} \quad \text{for } x \geq 0 $

## Poisson Distribution

### Probability Density Function (PDF):

$ P(X=k|\lambda) = \frac{e^{-\lambda}\lambda^k}{k!} \quad \text{for } k = 0, 1, 2, \ldots $

### Cumulative Distributive Function (CDF):

$ F(X=k|\lambda) = \sum_{i=0}^{k} \frac{e^{-\lambda}\lambda^i}{i!} \quad \text{for } k = 0, 1, 2, \ldots $

## Binomial Distribution

### Probability Density Function (PDF):
$ P(X=k|n,p) = \binom{n}{k} p^k (1-p)^{n-k} \quad \text{for } k = 0, 1, 2, \ldots, n $

### Cumulative Distributive Function (CDF):
$ F(X=k|n,p) = \sum_{i=0}^{k} \binom{n}{i} p^i (1-p)^{n-i} \quad \text{for } k = 0, 1, 2, \ldots, n $
