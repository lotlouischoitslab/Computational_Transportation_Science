```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Be careful here we are trying to find the minimum required thickness of subbase
data = pd.read_excel('data/subbase_data.xlsx')
data
```

```python
data.keys()
```

```python
sn = 2.50 # Design structural number 
```

```python
from sympy import *
x = symbols('x') 
data['Minimum Thickness (in)'][len(data['Minimum Thickness (in)'])-1] = x
data
```

```python
left = np.sum(data['Minimum Thickness (in)']*data['Strength Coefficient'])
right = sn 
equation = left - right 
solution = solve(equation,x)
thickness = solution[0]
print(f'Required thickness of subbase: {thickness} in')
```

```python
# Just look up the table 
front = 0.0877 
second = 0.658 

lef = front + 2*second 
lef = round(lef,3)
print(f'Load Equivalency Factor: {lef} lbs')
```
