```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
t = np.linspace(0,10)
m = 50
b = 10
y = m*t+b
y_list = []
```

```python
for i in range(10):
    y = m*(t-i)
    y_list.append(y)
```

```python

plt.xlabel('Time t')
plt.ylabel('Distance x')
plt.title('Trajectory Diagram')
for y in y_list:
    plt.plot(figsize=(10,20))
    plt.plot(t,y)
plt.gca().set_ylim(bottom=0)
```
